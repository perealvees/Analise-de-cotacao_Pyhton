```python
import yfinance
```

```python
cotacao = input("Digite o ticker desejado: ")

dados = yfinance.Ticker(cotacao).history("6mo")
fechamento = dados.Close
fechamento

fechamento.plot()
```

# trazer os valores com a minima, máxima e atual

```python
minima = fechamento.min()
```

```python
maxima = fechamento.max()
```

```python
fechamento
```

```python
atual = fechamento.iloc[-1]
```

```python
import pyautogui
import pyperclip
```

```python
pyautogui.PAUSE = 3
pyautogui.hotkey("crtl", "t")
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl","v")
pyautogui.hotkey("enter")
pyautogui.click(x=97, y=219)
pyperclip.copy("palvescosta@yahoo.com")
pyautogui.hotkey("ctrl","v")
pyautogui.hotkey("tab")
pyperclip.copy("Cotação do dia")
pyautogui.hotkey("ctrl","v")
pyautogui.hotkey("tab")
mensagem = f"""Bom dia!
Segue cotação da {cotacao} de hoje!
Cotação atual {round(atual, 3)} de hoje!
A máxima foi de {round(maxima, 3)} de hoje!
A mínima foi de {round(minima, 3)} de hoje!

Dúvidas estou a disposição!

"""
pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl","v")
pyautogui.click(x=865, y=694)
```
